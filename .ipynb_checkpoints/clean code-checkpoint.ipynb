GREENHOUSE GAS SCOPE DEFINITION:

```python
# === GHG EMISSION CATEGORIES AND GWP FACTORS ===
# Based on IPCC AR6 GWP100 values

# Define emission categories with their GWP factors
emission_categories = {
    # CO2 Emissions (GWP = 1.0)
    'CO2 - combustion - air': 1.0,
    'CO2 - waste - fossil - air': 1.0,
    'CO2 - agriculture - peat decay - air': 1.0,
    'CO2 - non combustion - Cement production - air': 1.0,
    'CO2 - non combustion - Lime production - air': 1.0,
    
    # CH4 Emissions - Combustion (GWP = 27.0)
    'CH4 - combustion - air': 27.0,
    
    # CH4 Emissions - Fugitive/Process (GWP = 29.8)
    'CH4 - non combustion - Extraction/production of (natural) gas - air': 29.8,
    'CH4 - non combustion - Extraction/production of crude oil - air': 29.8,
    'CH4 - non combustion - Oil refinery - air': 29.8,
    'CH4 - non combustion - Mining of bituminous coal - air': 29.8,
    'CH4 - non combustion - Mining of coking coal - air': 29.8,
    'CH4 - non combustion - Mining of lignite (brown coal) - air': 29.8,
    'CH4 - non combustion - Mining of sub-bituminous coal - air': 29.8,
    
    # CH4 Emissions - Biogenic (GWP = 27.0)
    'CH4 - agriculture - air': 27.0,
    'CH4 - waste - air': 27.0,
    
    # N2O Emissions (GWP = 273.0)
    'N2O - combustion - air': 273.0,
    'N2O - agriculture - air': 273.0,
    
    # Industrial GHGs
    'SF6 - air': 25184.0,
    'HFC - air': 1.0,  # Already in CO2-eq
    'PFC - air': 1.0   # Already in CO2-eq
}

print(f"Total emission categories included: {len(emission_categories)}")
print("\nEmission categories by type:")
for emission, gwp in emission_categories.items():
    print(f"  {emission} → GWP = {gwp}")
```

```python
# === VERIFY EMISSION AVAILABILITY ===
# Get air emissions data
air_emissions = ixi_data.air_emissions.F

# Check which emissions are available in the dataset
available_emissions = air_emissions.index.tolist()
print(f"\nTotal available air emissions in dataset: {len(available_emissions)}")

# Check if our required emissions are available
missing_emissions = []
available_required = []

for emission in emission_categories.keys():
    if emission in available_emissions:
        available_required.append(emission)
    else:
        missing_emissions.append(emission)

print(f"\nRequired emissions found: {len(available_required)}/{len(emission_categories)}")
if missing_emissions:
    print("Missing emissions:")
    for emission in missing_emissions:
        print(f"  - {emission}")
else:
    print("✓ All required emissions are available in the dataset!")
```

SINGLE YEAR ANALYSIS - PRODUCER APPROACH

```python
# === PRODUCER-BASED RESPONSIBILITY ===
print("=== CALCULATING PRODUCER-BASED RESPONSIBILITY ===")
print("Producer-based approach: Emissions allocated to where they physically occur")

# Extract emission data for our selected categories
total_ghg_emissions = np.zeros(len(air_emissions.columns))

print("\nCalculating GHG emissions by category:")
for emission_type, gwp_factor in emission_categories.items():
    if emission_type in available_emissions:
        # Get emissions for this type (in kg)
        emission_data = air_emissions.loc[emission_type].values
        
        # Convert to CO2-equivalent tons
        co2_equiv_tons = (emission_data * gwp_factor) / 1000  # kg to tons
        
        # Add to total
        total_ghg_emissions += co2_equiv_tons
        
        # Report
        total_emission = co2_equiv_tons.sum()
        print(f"  {emission_type[:50]:<50}: {total_emission:>12,.0f} tons CO2-eq")

# Create producer responsibility dataframe
producer_responsibility = pd.Series(
    total_ghg_emissions, 
    index=air_emissions.columns,
    name='Producer_Responsibility_tons_CO2eq'
)

print(f"\n=== PRODUCER-BASED RESULTS ===")
print(f"Total global GHG emissions: {producer_responsibility.sum():,.0f} tons CO2-eq")
print(f"Average per sector: {producer_responsibility.mean():,.0f} tons CO2-eq")
print(f"Maximum single sector: {producer_responsibility.max():,.0f} tons CO2-eq")
```

```python
# === PRODUCER RESPONSIBILITY BY COUNTRY ===
# Extract country codes from multi-index
countries = [sector_region[0] for sector_region in producer_responsibility.index]
country_producer_responsibility = pd.Series(
    producer_responsibility.values, 
    index=countries
).groupby(level=0).sum()

# Sort by emissions (highest first)
country_producer_responsibility = country_producer_responsibility.sort_values(ascending=False)

print("=== TOP 10 COUNTRIES - PRODUCER RESPONSIBILITY ===")
print("(Emissions occurring within national borders)")
for i, (country, emissions) in enumerate(country_producer_responsibility.head(10).items(), 1):
    percentage = (emissions / country_producer_responsibility.sum()) * 100
    print(f"{i:2d}. {country}: {emissions:>12,.0f} tons CO2-eq ({percentage:5.1f}%)")

# Verification
total_by_country = country_producer_responsibility.sum()
total_by_sector = producer_responsibility.sum()
print(f"\nVerification:")
print(f"Sum by country: {total_by_country:,.0f} tons CO2-eq")
print(f"Sum by sector:  {total_by_sector:,.0f} tons CO2-eq")
print(f"Difference: {abs(total_by_country - total_by_sector):,.0f} tons CO2-eq ({'✓ Match' if abs(total_by_country - total_by_sector) < 1 else '✗ Mismatch'})")
```

CONSUMER bASED RESPONSABILITY

```python
# === CONSUMER-BASED RESPONSIBILITY ===
print("=== CALCULATING CONSUMER-BASED RESPONSIBILITY ===")
print("Consumer-based approach: Emissions allocated to final consumers of goods/services")
print("Formula: t' = f' × B × ŷ (Piñero et al., 2019)")
```

```python
print("\n--- Step 1: Calculate Emission Intensities (f') ---")

# Get total economic output by sector
total_output = ixi_data.x.values.flatten()
print(f"Total output vector shape: {total_output.shape}")
print(f"Global economic output: {total_output.sum():,.0f} million EUR")

# Calculate emission intensity (tons CO2-eq per million EUR output)
emission_intensity = np.divide(
    producer_responsibility.values, 
    total_output, 
    out=np.zeros_like(producer_responsibility.values), 
    where=(total_output != 0)
)

# Check for zero output sectors
zero_output_sectors = np.sum(total_output == 0)
positive_output_mask = total_output > 0

print(f"Emission intensity vector shape: {emission_intensity.shape}")
print(f"Sectors with zero output: {zero_output_sectors} ({zero_output_sectors/len(total_output)*100:.1f}%)")

if np.sum(positive_output_mask) > 0:
    avg_intensity = emission_intensity[positive_output_mask].mean()
    max_intensity = emission_intensity[positive_output_mask].max()
    print(f"Average emission intensity (positive sectors): {avg_intensity:,.2f} tons CO2-eq/million EUR")
    print(f"Max emission intensity: {max_intensity:,.2f} tons CO2-eq/million EUR")

print("\nInterpretation: f'[i] = tons CO2-eq emitted per million EUR of output from sector i")
```

```python
print("\n--- Step 1: Calculate Emission Intensities (f') ---")

# Get total economic output by sector
total_output = ixi_data.x.values.flatten()
print(f"Total output vector shape: {total_output.shape}")
print(f"Global economic output: {total_output.sum():,.0f} million EUR")

# Calculate emission intensity (tons CO2-eq per million EUR output)
# Note: Not handling zero output sectors for now
emission_intensity = producer_responsibility.values / total_output

# Replace any infinite or NaN values with zero
emission_intensity = np.where(np.isfinite(emission_intensity), emission_intensity, 0)

print(f"Emission intensity vector shape: {emission_intensity.shape}")
print(f"Average emission intensity: {emission_intensity[emission_intensity > 0].mean():,.2f} tons CO2-eq/million EUR")

print("\nInterpretation: f'[i] = tons CO2-eq emitted per million EUR of output from sector i")
```

```python
print("\n--- Step 2: Calculate Leontief Inverse Matrix (B) ---")

# Calculate all system matrices using pymrio
ixi_data.calc_system()
print("System matrices calculated successfully")

# Get the Leontief inverse matrix (this is our B matrix from Piñero's formula)
B_matrix = ixi_data.L
print(f"Leontief matrix (B) shape: {B_matrix.shape}")

print("\nInterpretation: B[i,j] = total output from sector i required")
print("                        (directly + indirectly) per unit of final demand for sector j")
```

```python
print("\n--- Step 3: Prepare Final Demand Matrix (Y) ---")

# Get final demand matrix (sectors × demand categories)
final_demand = ixi_data.Y
print(f"Final demand matrix shape: {final_demand.shape}")

# Show final demand categories
print(f"\nFinal demand categories ({final_demand.shape[1]} total):")
demand_categories = final_demand.columns.tolist()
for i, category in enumerate(demand_categories[:10], 1):
    print(f"  {i:2d}. {category}")
if len(demand_categories) > 10:
    print(f"  ... and {len(demand_categories) - 10} more categories")

print(f"\nTotal global final demand: {final_demand.sum().sum():,.0f} million EUR")
print("\nInterpretation: Y = final demand matrix (sectors × demand categories × regions)")
```

```python
print("\n--- Step 4: Calculate Consumer Responsibility (Full Matrix) ---")
print("Applying full Piñero formula: t' = f' × B × Y")

# Debug matrix shapes first
print(f"Emission intensity shape: {emission_intensity.shape}")
print(f"B matrix shape: {B_matrix.shape}")
print(f"Final demand shape: {final_demand.shape}")

# Calculate intermediate step first
print("Step 4a: Calculate total emission intensity (f' × B)")
total_emission_intensity = emission_intensity @ B_matrix.values
print(f"Total emission intensity shape: {total_emission_intensity.shape}")

# Step 4b: Apply to final demand
print("Step 4b: Apply to final demand by region")
final_demand_by_region = final_demand.sum(axis=1)  # Sum across final demand categories
print(f"Final demand by region shape: {final_demand_by_region.shape}")

# Calculate consumer responsibility (should be a vector)
consumer_responsibility = total_emission_intensity * final_demand_by_region.values
print(f"Consumer responsibility shape: {consumer_responsibility.shape}")

# Also calculate the full detailed version for category analysis
print("Step 4c: Full category breakdown")
consumer_by_category_matrix = np.outer(total_emission_intensity, np.ones(final_demand.shape[1])) * final_demand.values
consumer_by_category_totals = consumer_by_category_matrix.sum(axis=0)  # Sum across sectors
print(f"Consumer by category shape: {consumer_by_category_totals.shape}")
```

```python
print("\n--- Step 5: Create Detailed Breakdown ---")

# Create breakdown by final demand category (global totals)
consumer_by_category = pd.Series(
    consumer_by_category_totals,
    index=final_demand.columns,
    name='Consumer_Responsibility_by_Category'
)

print("=== GLOBAL RESPONSIBILITY BY FINAL DEMAND CATEGORY ===")
for i, (category, emissions) in enumerate(consumer_by_category.head(10).items(), 1):
    percentage = (emissions / consumer_by_category.sum()) * 100
    category_str = str(category)[:60] if not isinstance(category, str) else category[:60]
    print(f"{i:2d}. {category_str:<60}: {emissions:>12,.0f} tons CO2-eq ({percentage:5.1f}%)")

print(f"\nTotal consumer responsibility shape: {consumer_responsibility.shape}")
```

```python
print("\n=== CONSUMER-BASED RESULTS & VERIFICATION ===")

# Mass conservation check
producer_total = producer_responsibility.sum()
consumer_total = consumer_responsibility.sum()
conservation_ratio = consumer_total / producer_total

print(f"Producer responsibility total:  {producer_total:,.0f} tons CO2-eq")
print(f"Consumer responsibility total:  {consumer_total:,.0f} tons CO2-eq")
print(f"Mass conservation ratio:       {conservation_ratio:.6f}")
print(f"Conservation quality:          {'✓ Excellent' if abs(1 - conservation_ratio) < 0.001 else '✓ Good' if abs(1 - conservation_ratio) < 0.01 else '⚠ Check required'}")

# Create consumer responsibility series with proper index
consumer_responsibility_series = pd.Series(
    consumer_responsibility,
    index=final_demand_by_region.index,
    name='Consumer_Responsibility_tons_CO2eq'
)
```

```python
# === CONSUMER RESPONSIBILITY BY COUNTRY ===
# Extract country codes from multi-index  
countries = [sector_region[0] for sector_region in consumer_responsibility_series.index]
country_consumer_responsibility = pd.Series(
    consumer_responsibility_series.values,
    index=countries
).groupby(level=0).sum()

# Sort by emissions (highest first)
country_consumer_responsibility = country_consumer_responsibility.sort_values(ascending=False)

print("\n=== TOP 10 COUNTRIES - CONSUMER RESPONSIBILITY ===")
print("(Emissions embedded in final consumption)")
for i, (country, emissions) in enumerate(country_consumer_responsibility.head(10).items(), 1):
    percentage = (emissions / country_consumer_responsibility.sum()) * 100
    print(f"{i:2d}. {country}: {emissions:>12,.0f} tons CO2-eq ({percentage:5.1f}%)")

print(f"\n=== ECONOMIC INTERPRETATION ===")
print("Example: If USA imports German cars:")
print("  • Producer approach: Germany responsible for car manufacturing emissions")
print("  • Consumer approach: USA responsible for emissions embedded in cars they consume")
```

```python
# === VALUE-ADDED-BASED RESPONSIBILITY ===
print("=== CALCULATING VALUE-ADDED-BASED RESPONSIBILITY ===")
print("Value-added approach: Emissions allocated based on economic value captured in supply chains")
print("Formula: h = v̂ × B × t (Piñero et al., 2019)")
print("Where: h = value-added responsibility, v̂ = value-added coefficients, B = Leontief inverse, t = consumer responsibility")
```

Extract Value-Added Data

```python
print("\n--- Step 1: Extract Value-Added Data ---")

# Get factor inputs (value-added components) from EXIOBASE3
factor_inputs = ixi_data.factor_inputs.F
print(f"Factor inputs shape: {factor_inputs.shape}")

# Display available value-added components
print("Available value-added components:")
for i, component in enumerate(factor_inputs.index, 1):
    print(f"  {i:2d}. {component}")
```

Define and Calculate Total Value-Added

```python
print("\n--- Step 2: Calculate Total Value-Added ---")

# Define value-added components (following standard national accounting)
value_added_components = [
    # Labor compensation
    "Compensation of employees; wages, salaries, & employers' social contributions: Low-skilled",
    "Compensation of employees; wages, salaries, & employers' social contributions: Medium-skilled", 
    "Compensation of employees; wages, salaries, & employers' social contributions: High-skilled",
    
    # Capital returns
    'Operating surplus: Consumption of fixed capital',       # Depreciation
    'Operating surplus: Rents on land',                      # Land rents
    'Operating surplus: Royalties on resources',             # Resource royalties
    'Operating surplus: Remaining net operating surplus',    # Profits
    
    # Government share
    'Other net taxes on production'  # Net production taxes
]

print("Value-added components included:")
for i, component in enumerate(value_added_components, 1):
    print(f"  {i}. {component}")

# Calculate total value-added by sector
available_components = [comp for comp in value_added_components if comp in factor_inputs.index]
missing_components = [comp for comp in value_added_components if comp not in factor_inputs.index]

if missing_components:
    print(f"\nWarning: {len(missing_components)} components not found in data:")
    for comp in missing_components:
        print(f"  - {comp}")

print(f"\nUsing {len(available_components)} available components")

# Sum value-added components
total_value_added = factor_inputs.loc[available_components].sum(axis=0)
print(f"Total value-added shape: {total_value_added.shape}")
print(f"Global value-added: {total_value_added.sum():,.0f} million EUR")
```

Calculate Value-Added Coefficients

```python
print("\n--- Step 3: Calculate Value-Added Coefficients (v̂) ---")

# Get total economic output (same as before)
total_output = ixi_data.x.values.flatten()

# Calculate value-added coefficients (value-added per unit of output)
va_coefficients = np.divide(
    total_value_added.values, 
    total_output, 
    out=np.zeros_like(total_value_added.values), 
    where=(total_output != 0)
)

# Data quality cleaning (following methodology document)
print("Data quality cleaning:")

# Handle negative value-added (data quality issue)
negative_va = np.sum(va_coefficients < 0)
if negative_va > 0:
    print(f"  Found {negative_va} sectors with negative value-added (set to minimum €1M)")
    va_coefficients = np.where(va_coefficients < 0, 1.0/total_output.max(), va_coefficients)

# Cap value-added coefficients at 1.0 (value-added cannot exceed total output)
above_one = np.sum(va_coefficients > 1.0)
if above_one > 0:
    print(f"  Found {above_one} sectors with VA coefficient > 1.0 (capped at 1.0)")
    va_coefficients = np.minimum(va_coefficients, 1.0)

# Check for zero output sectors
zero_output = np.sum(total_output == 0)
print(f"  Sectors with zero output: {zero_output} (assigned VA coefficient = 0)")

print(f"Value-added coefficients shape: {va_coefficients.shape}")
print(f"Average VA coefficient: {va_coefficients[va_coefficients > 0].mean():.4f}")
print(f"Max VA coefficient: {va_coefficients.max():.4f}")

print("\nInterpretation: v̂[i] = value-added created per EUR of output in sector i")
```

Apply pineros Formula

```python
print("\n--- Step 4: Apply Piñero's Formula h = v̂ × B × t ---")

# Create diagonal matrix of value-added coefficients
va_coeff_diag = np.diag(va_coefficients)
print(f"Value-added diagonal matrix shape: {va_coeff_diag.shape}")

# Apply Piñero's formula: h = v̂ × B × t
# Where t is the consumer responsibility vector from previous step
print("Calculating: va_coefficients_diag @ B_matrix @ consumer_responsibility_series")

value_added_responsibility = va_coeff_diag @ B_matrix.values @ consumer_responsibility_series.values
print(f"Value-added responsibility shape: {value_added_responsibility.shape}")

print("\nEconomic interpretation:")
print("This reallocates consumer-based emissions according to who captures")
print("economic value (wages, profits, rents) in the supply chain")
```

Step 5: Mass Conservation Normalization

```python
print("\n--- Step 5: Mass Conservation Normalization ---")

# Calculate normalization factor for perfect mass conservation
consumer_total = consumer_responsibility_series.sum()
va_total_before = value_added_responsibility.sum()
normalization_factor = consumer_total / va_total_before

print(f"Consumer responsibility total:     {consumer_total:,.0f} tons CO2-eq")
print(f"VA responsibility (before norm):   {va_total_before:,.0f} tons CO2-eq")
print(f"Normalization factor:             {normalization_factor:.6f}")

# Apply normalization
value_added_responsibility_normalized = value_added_responsibility * normalization_factor
va_total_after = value_added_responsibility_normalized.sum()

print(f"VA responsibility (after norm):    {va_total_after:,.0f} tons CO2-eq")
print(f"Conservation check:               {'✓ Perfect' if abs(va_total_after - consumer_total) < 1 else '⚠ Check required'}")

# Use normalized values
value_added_responsibility = value_added_responsibility_normalized

print("\nWhy normalize: Ensures total emissions are conserved (academic rigor requirement)")
```

Results and country analysis

```python
print("\n=== VALUE-ADDED-BASED RESULTS ===")

# Create value-added responsibility series
value_added_responsibility_series = pd.Series(
    value_added_responsibility,
    index=consumer_responsibility_series.index,
    name='ValueAdded_Responsibility_tons_CO2eq'
)

# Value-added responsibility by country
countries = [sector_region[0] for sector_region in value_added_responsibility_series.index]
country_va_responsibility = pd.Series(
    value_added_responsibility_series.values,
    index=countries
).groupby(level=0).sum()

# Sort by emissions (highest first)
country_va_responsibility = country_va_responsibility.sort_values(ascending=False)

print("=== TOP 10 COUNTRIES - VALUE-ADDED RESPONSIBILITY ===")
print("(Emissions allocated based on value captured in supply chains)")
for i, (country, emissions) in enumerate(country_va_responsibility.head(10).items(), 1):
    percentage = (emissions / country_va_responsibility.sum()) * 100
    print(f"{i:2d}. {country}: {emissions:>12,.0f} tons CO2-eq ({percentage:5.1f}%)")

print(f"\n=== ECONOMIC INTERPRETATION ===")
print("Example: Global smartphone supply chain:")
print("  • Producer approach: China responsible (where phones manufactured)")
print("  • Consumer approach: USA responsible (where phones consumed)")
print("  • Value-added approach:")
print("    - USA responsible for design/software (high value-added)")
print("    - Germany responsible for machinery/engineering (high value-added)")
print("    - China responsible for assembly (lower value-added)")
```

```python
print("=== COMPREHENSIVE THREE-WAY COMPARISON ===")

# Combine all three approaches - make sure we include ALL regions
comparison_df = pd.DataFrame({
    'Producer': country_producer_responsibility,
    'Consumer': country_consumer_responsibility,
    'Value_Added': country_va_responsibility
}).fillna(0)

# Sort by total emissions (sum across all approaches) to get the real top emitters
comparison_df['Total'] = comparison_df.sum(axis=1)
comparison_df = comparison_df.sort_values('Total', ascending=False)

# Calculate percentage shares
comparison_pct = comparison_df[['Producer', 'Consumer', 'Value_Added']].div(
    comparison_df[['Producer', 'Consumer', 'Value_Added']].sum(axis=0), axis=1
) * 100

print("=== TOP 15 REGIONS - ALL THREE APPROACHES ===")
print("Rank | Region | Producer (%) | Consumer (%) | Value-Added (%) | Biggest Shift")
print("-" * 85)

for i, country in enumerate(comparison_pct.index[:15], 1):
    prod_pct = comparison_pct.loc[country, 'Producer']
    cons_pct = comparison_pct.loc[country, 'Consumer']
    va_pct = comparison_pct.loc[country, 'Value_Added']
    
    # Find biggest difference
    max_val = max(prod_pct, cons_pct, va_pct)
    min_val = min(prod_pct, cons_pct, va_pct)
    biggest_shift = max_val - min_val
    
    print(f"{i:2d}   | {country:6s} | {prod_pct:8.1f}     | {cons_pct:8.1f}     | {va_pct:10.1f}      | {biggest_shift:6.1f}pp")

# Also show the actual emission values for context
print(f"\n=== TOP 10 REGIONS - ABSOLUTE VALUES (billion tons CO2-eq) ===")
print("Rank | Region | Producer | Consumer | Value-Added")
print("-" * 55)

comparison_abs = comparison_df[['Producer', 'Consumer', 'Value_Added']] / 1e9  # Convert to billion tons

for i, country in enumerate(comparison_abs.index[:10], 1):
    prod_abs = comparison_abs.loc[country, 'Producer']
    cons_abs = comparison_abs.loc[country, 'Consumer']
    va_abs = comparison_abs.loc[country, 'Value_Added']
    
    print(f"{i:2d}   | {country:6s} | {prod_abs:8.1f} | {cons_abs:8.1f} | {va_abs:9.1f}")
```

```python
print("\n=== RESPONSIBILITY SHIFTS ANALYSIS ===")

# Calculate differences between approaches
comparison_df['Prod_to_Cons'] = comparison_pct['Consumer'] - comparison_pct['Producer']
comparison_df['Cons_to_VA'] = comparison_pct['Value_Added'] - comparison_pct['Consumer']
comparison_df['Prod_to_VA'] = comparison_pct['Value_Added'] - comparison_pct['Producer']

print("BIGGEST WINNERS (gain responsibility):")
print("From Producer to Consumer approach:")
winners_pc = comparison_df['Prod_to_Cons'].sort_values(ascending=False).head(5)
for country, gain in winners_pc.items():
    print(f"  {country}: +{gain:.1f} percentage points")

print("\nFrom Consumer to Value-Added approach:")
winners_cv = comparison_df['Cons_to_VA'].sort_values(ascending=False).head(5)
for country, gain in winners_cv.items():
    print(f"  {country}: +{gain:.1f} percentage points")

print("\nBIGGEST LOSERS (lose responsibility):")
print("From Producer to Consumer approach:")
losers_pc = comparison_df['Prod_to_Cons'].sort_values(ascending=True).head(5)
for country, loss in losers_pc.items():
    print(f"  {country}: {loss:.1f} percentage points")

print("\nFrom Consumer to Value-Added approach:")
losers_cv = comparison_df['Cons_to_VA'].sort_values(ascending=True).head(5)
for country, loss in losers_cv.items():
    print(f"  {country}: {loss:.1f} percentage points")
```

```python
# === SECTION 4: TIME SERIES ANALYSIS (1995-2022) ===
print("=== TIME SERIES ANALYSIS FOR ALL THREE APPROACHES ===")
print("Analyzing responsibility allocation trends from 1995-2022")

import time
import pickle
from datetime import datetime
from pathlib import Path

# === PICKLE CONFIGURATION ===
RESULTS_FOLDER = Path(r"C:\Users\Marine Riesterer\OneDrive\Desktop\MA Arbeit\Results")
RESULTS_FOLDER.mkdir(exist_ok=True)  # Create folder if it doesn't exist

# Pickle file names
PICKLE_FILE_SAMPLE = RESULTS_FOLDER / "time_series_results_sample.pkl"
PICKLE_FILE_FULL = RESULTS_FOLDER / "time_series_results_full.pkl"

# === EMISSION CATEGORIES DEFINITION ===
emission_categories = {
    # CO2 Emissions (GWP = 1.0)
    'CO2 - combustion - air': 1.0,
    'CO2 - waste - fossil - air': 1.0,
    'CO2 - agriculture - peat decay - air': 1.0,
    'CO2 - non combustion - Cement production - air': 1.0,
    'CO2 - non combustion - Lime production - air': 1.0,
    
    # CH4 Emissions - Combustion (GWP = 27.0)
    'CH4 - combustion - air': 27.0,
    
    # CH4 Emissions - Fugitive/Process (GWP = 29.8)
    'CH4 - non combustion - Extraction/production of (natural) gas - air': 29.8,
    'CH4 - non combustion - Extraction/production of crude oil - air': 29.8,
    'CH4 - non combustion - Oil refinery - air': 29.8,
    'CH4 - non combustion - Mining of bituminous coal - air': 29.8,
    'CH4 - non combustion - Mining of coking coal - air': 29.8,
    'CH4 - non combustion - Mining of lignite (brown coal) - air': 29.8,
    'CH4 - non combustion - Mining of sub-bituminous coal - air': 29.8,
    
    # CH4 Emissions - Biogenic (GWP = 27.0)
    'CH4 - agriculture - air': 27.0,
    'CH4 - waste - air': 27.0,
    
    # N2O Emissions (GWP = 273.0)
    'N2O - combustion - air': 273.0,
    'N2O - agriculture - air': 273.0,
    
    # Industrial GHGs
    'SF6 - air': 25184.0,
    'HFC - air': 1.0,  # Already in CO2-eq
    'PFC - air': 1.0   # Already in CO2-eq
}

# === VALUE-ADDED COMPONENTS DEFINITION ===
value_added_components = [
    "Compensation of employees; wages, salaries, & employers' social contributions: Low-skilled",
    "Compensation of employees; wages, salaries, & employers' social contributions: Medium-skilled", 
    "Compensation of employees; wages, salaries, & employers' social contributions: High-skilled",
    'Operating surplus: Consumption of fixed capital',
    'Operating surplus: Rents on land',
    'Operating surplus: Royalties on resources',
    'Operating surplus: Remaining net operating surplus',
    'Other net taxes on production'
]

# Analysis parameters
ANALYSIS_YEARS = list(range(1995, 2023))  # 1995-2022 (28 years)
SAMPLE_YEARS = [1995, 2000, 2005, 2010, 2015, 2020, 2022]  # For testing

# Choose analysis mode
USE_SAMPLE = False  # Set to True for sample, False for full analysis
FORCE_RERUN = False  # Set to True to force recalculation even if pickle exists

years_to_analyze = SAMPLE_YEARS if USE_SAMPLE else ANALYSIS_YEARS
pickle_file = PICKLE_FILE_SAMPLE if USE_SAMPLE else PICKLE_FILE_FULL

print(f"Analysis mode: {'Sample' if USE_SAMPLE else 'Full'}")
print(f"Years to analyze: {len(years_to_analyze)} years ({years_to_analyze[0]}-{years_to_analyze[-1]})")
print(f"Pickle file: {pickle_file}")

# Data path
ixi_folder = Path(r"C:\Users\Marine Riesterer\OneDrive\Desktop\MA Arbeit\Exiobase3_ixi_latest")
```

```python
# === ANALYSIS FUNCTION DEFINITION ===
def analyze_single_year_complete(year):
    """
    Complete three-approach analysis for one year
    Returns: producer_country, consumer_country, va_country, global_totals, processing_time, quality_info
    """
    start_time = time.time()
    
    try:
        print(f"\n--- Processing {year} ---")
        
        # Load data
        ixi_data = pymrio.parse_exiobase3(str(ixi_folder / f"IOT_{year}_ixi.zip"))
        print(f"  ✓ Data loaded: {year}")
        
        # === PRODUCER-BASED APPROACH ===
        air_emissions = ixi_data.air_emissions.F
        
        # Check available emissions for this year
        available_emissions = air_emissions.index.tolist()
        missing_emissions = [em for em in emission_categories.keys() if em not in available_emissions]
        
        # Calculate total GHG emissions
        total_ghg_emissions = np.zeros(len(air_emissions.columns))
        
        for emission_type, gwp_factor in emission_categories.items():
            if emission_type in available_emissions:
                emission_data = air_emissions.loc[emission_type].values
                co2_equiv_tons = (emission_data * gwp_factor) / 1000  # kg to tons
                total_ghg_emissions += co2_equiv_tons
        
        # Producer responsibility by country
        producer_responsibility = pd.Series(total_ghg_emissions, index=air_emissions.columns)
        countries = [sector_region[0] for sector_region in producer_responsibility.index]
        country_producer = pd.Series(producer_responsibility.values, index=countries).groupby(level=0).sum()
        print(f"  ✓ Producer approach: {country_producer.sum()/1e9:.1f} Gt CO2-eq")
        
        # === CONSUMER-BASED APPROACH ===
        # Calculate system matrices
        ixi_data.calc_system()
        
        # Emission intensities (handle division by zero)
        total_output = ixi_data.x.values.flatten()
        emission_intensity = np.divide(
            producer_responsibility.values, 
            total_output, 
            out=np.zeros_like(producer_responsibility.values), 
            where=(total_output != 0)
        )
        
        # Consumer responsibility calculation
        B_matrix = ixi_data.L
        final_demand = ixi_data.Y
        final_demand_by_region = final_demand.sum(axis=1)
        
        total_emission_intensity = emission_intensity @ B_matrix.values
        consumer_responsibility = total_emission_intensity * final_demand_by_region.values
        
        # Consumer responsibility by country
        consumer_countries = [sector_region[0] for sector_region in final_demand_by_region.index]
        country_consumer = pd.Series(consumer_responsibility, index=consumer_countries).groupby(level=0).sum()
        print(f"  ✓ Consumer approach: {country_consumer.sum()/1e9:.1f} Gt CO2-eq")
        
        # === VALUE-ADDED-BASED APPROACH ===
        factor_inputs = ixi_data.factor_inputs.F
        
        # Check available value-added components for this year
        available_va_components = [comp for comp in value_added_components if comp in factor_inputs.index]
        missing_va_components = [comp for comp in value_added_components if comp not in factor_inputs.index]
        
        if len(available_va_components) == 0:
            print(f"  ⚠ WARNING: No value-added components found for {year}")
            return None, None, None, None, 0, None
        
        total_value_added = factor_inputs.loc[available_va_components].sum(axis=0)
        
        # Value-added coefficients (handle division by zero)
        va_coefficients = np.divide(
            total_value_added.values, 
            total_output, 
            out=np.zeros_like(total_value_added.values), 
            where=(total_output != 0)
        )
        
        # Data quality cleaning
        va_coefficients = np.where(va_coefficients < 0, 1.0/total_output.max(), va_coefficients)
        va_coefficients = np.minimum(va_coefficients, 1.0)
        
        # Value-added responsibility calculation
        va_coeff_diag = np.diag(va_coefficients)
        consumer_responsibility_series = pd.Series(consumer_responsibility, index=final_demand_by_region.index)
        
        value_added_responsibility = va_coeff_diag @ B_matrix.values @ consumer_responsibility_series.values
        
        # Mass conservation normalization
        if value_added_responsibility.sum() > 0:
            normalization_factor = consumer_responsibility.sum() / value_added_responsibility.sum()
            value_added_responsibility = value_added_responsibility * normalization_factor
        
        # Value-added responsibility by country
        va_countries = [sector_region[0] for sector_region in consumer_responsibility_series.index]
        country_va = pd.Series(value_added_responsibility, index=va_countries).groupby(level=0).sum()
        print(f"  ✓ Value-added approach: {country_va.sum()/1e9:.1f} Gt CO2-eq")
        
        # Global totals
        global_totals = {
            'producer': country_producer.sum(),
            'consumer': country_consumer.sum(), 
            'value_added': country_va.sum()
        }
        
        # Data quality info
        quality_info = {
            'missing_emissions': missing_emissions,
            'missing_va_components': missing_va_components,
            'available_emissions': len(available_emissions),
            'available_va_components': len(available_va_components)
        }
        
        processing_time = time.time() - start_time
        print(f"  ✓ Completed {year} in {processing_time:.1f} seconds")
        
        return country_producer, country_consumer, country_va, global_totals, processing_time, quality_info
        
    except Exception as e:
        print(f"  ✗ ERROR processing {year}: {str(e)}")
        return None, None, None, None, 0, None
```

```python
# === PICKLE FUNCTIONS ===
def save_time_series_results(results, pickle_file):
    """Save time series results to pickle file"""
    try:
        with open(pickle_file, 'wb') as f:
            pickle.dump(results, f)
        print(f"✓ Results saved to: {pickle_file}")
        print(f"  File size: {pickle_file.stat().st_size / 1024 / 1024:.1f} MB")
    except Exception as e:
        print(f"✗ Error saving results: {e}")

def load_time_series_results(pickle_file):
    """Load time series results from pickle file"""
    try:
        with open(pickle_file, 'rb') as f:
            results = pickle.load(f)
        print(f"✓ Results loaded from: {pickle_file}")
        print(f"  Years: {len(results['years'])}")
        print(f"  Countries: {len(results['producer_country'])}")
        return results
    except FileNotFoundError:
        print(f"✗ Pickle file not found: {pickle_file}")
        return None
    except Exception as e:
        print(f"✗ Error loading results: {e}")
        return None

def check_results_completeness(results, expected_years):
    """Check if loaded results match expected years"""
    if results is None:
        return False
    
    loaded_years = set(results['years'])
    expected_years_set = set(expected_years)
    
    missing_years = expected_years_set - loaded_years
    
    if missing_years:
        print(f"✗ Missing years in saved results: {sorted(missing_years)}")
        return False
    
    print(f"✓ Results complete for all {len(expected_years)} requested years")
    return True
```

```python
# === MAIN ANALYSIS EXECUTION ===
print(f"\n=== CHECKING FOR EXISTING RESULTS ===")

# Try to load existing results
time_series_results = None

if not FORCE_RERUN:
    time_series_results = load_time_series_results(pickle_file)
    
    if time_series_results is not None:
        # Check if results are complete
        if check_results_completeness(time_series_results, years_to_analyze):
            print("✓ Using existing results - no need to rerun analysis")
            print(f"  Analysis was run on: {time_series_results.get('analysis_date', 'Unknown date')}")
        else:
            print("⚠ Existing results incomplete - will rerun analysis")
            time_series_results = None

# If no valid results found, run the analysis
if time_series_results is None or FORCE_RERUN:
    print(f"\n=== RUNNING TIME SERIES ANALYSIS ===")
    print(f"Starting analysis at {datetime.now().strftime('%H:%M:%S')}")
    print(f"Estimated time: {len(years_to_analyze) * 0.5:.0f}-{len(years_to_analyze) * 2:.0f} minutes")
    
    # Initialize storage for results
    time_series_results = {
        'years': [],
        'producer_country': {},
        'consumer_country': {},
        'value_added_country': {},
        'global_totals': {'producer': [], 'consumer': [], 'value_added': []},
        'processing_times': [],
        'data_quality': {'missing_emissions': [], 'missing_va_components': []},
        'analysis_date': datetime.now().isoformat(),
        'analysis_mode': 'Sample' if USE_SAMPLE else 'Full',
        'years_analyzed': years_to_analyze
    }
    
    # Initialize storage for all countries across all years
    all_countries = set()
    
    # Run analysis for each year
    for i, year in enumerate(years_to_analyze, 1):
        print(f"\nProcessing {i}/{len(years_to_analyze)}: {year}")
        
        result = analyze_single_year_complete(year)
        country_producer, country_consumer, country_va, global_totals, proc_time, quality_info = result
        
        if country_producer is not None:  # Successful processing
            # Store results
            time_series_results['years'].append(year)
            time_series_results['global_totals']['producer'].append(global_totals['producer'])
            time_series_results['global_totals']['consumer'].append(global_totals['consumer'])
            time_series_results['global_totals']['value_added'].append(global_totals['value_added'])
            time_series_results['processing_times'].append(proc_time)
            
            # Store data quality info
            if quality_info:
                time_series_results['data_quality']['missing_emissions'].append((year, quality_info['missing_emissions']))
                time_series_results['data_quality']['missing_va_components'].append((year, quality_info['missing_va_components']))
            
            # Store country data
            all_year_countries = set(country_producer.index) | set(country_consumer.index) | set(country_va.index)
            all_countries.update(all_year_countries)
            
            for country in all_year_countries:
                if country not in time_series_results['producer_country']:
                    time_series_results['producer_country'][country] = []
                    time_series_results['consumer_country'][country] = []
                    time_series_results['value_added_country'][country] = []
                
                time_series_results['producer_country'][country].append(country_producer.get(country, 0))
                time_series_results['consumer_country'][country].append(country_consumer.get(country, 0))
                time_series_results['value_added_country'][country].append(country_va.get(country, 0))
        
        # Save intermediate results every 5 years (for long runs)
        if i % 5 == 0:
            save_time_series_results(time_series_results, pickle_file)
            print(f"  💾 Intermediate save completed")
    
    print(f"\n=== TIME SERIES ANALYSIS COMPLETED ===")
    print(f"Successfully processed: {len(time_series_results['years'])} years")
    print(f"Total countries tracked: {len(all_countries)}")
    if time_series_results['processing_times']:
        print(f"Average processing time: {np.mean(time_series_results['processing_times']):.1f} seconds/year")
        print(f"Total processing time: {sum(time_series_results['processing_times'])/60:.1f} minutes")
    
    # Final save
    print(f"\n=== SAVING FINAL RESULTS ===")
    save_time_series_results(time_series_results, pickle_file)

else:
    print(f"\n=== USING CACHED RESULTS ===")

print(f"\n✓ Time series results ready for analysis!")
print(f"  Years available: {time_series_results['years']}")
print(f"  Countries tracked: {len(time_series_results['producer_country'])}")
```

```python
# === COMPREHENSIVE TIME SERIES VISUALIZATIONS ===
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set professional plotting style
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300

def create_comprehensive_time_series_plots(time_series_results):
    """Create comprehensive time series visualizations"""
    
    years = time_series_results['years']
    
    # Create main figure with subplots
    fig = plt.figure(figsize=(18, 12))
    
    # === PLOT 1: GLOBAL EMISSIONS TRENDS ===
    ax1 = plt.subplot(2, 3, 1)
    
    producer_vals = [x/1e9 for x in time_series_results['global_totals']['producer']]
    consumer_vals = [x/1e9 for x in time_series_results['global_totals']['consumer']]
    va_vals = [x/1e9 for x in time_series_results['global_totals']['value_added']]
    
    ax1.plot(years, producer_vals, 'o-', linewidth=2.5, markersize=6, label='Producer-based', color='#e74c3c')
    ax1.plot(years, consumer_vals, 's-', linewidth=2.5, markersize=6, label='Consumer-based', color='#3498db')
    ax1.plot(years, va_vals, '^-', linewidth=2.5, markersize=6, label='Value-added-based', color='#2ecc71')
    
    ax1.set_xlabel('Year', fontsize=11)
    ax1.set_ylabel('Global GHG Emissions (Gt CO₂-eq)', fontsize=11)
    ax1.set_title('Global GHG Emissions Trends\n(All Three Approaches)', fontsize=12, fontweight='bold')
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(bottom=0)
    
    # Add trend lines
    if len(years) > 3:
        z_prod = np.polyfit(years, producer_vals, 1)
        p_prod = np.poly1d(z_prod)
        ax1.plot(years, p_prod(years), "--", alpha=0.7, color='#e74c3c', linewidth=1)
    
    # === PLOT 2: APPROACH DIFFERENCES ===
    ax2 = plt.subplot(2, 3, 2)
    
    # Calculate differences from producer approach
    cons_diff = [(c-p)/p*100 for c, p in zip(consumer_vals, producer_vals)]
    va_diff = [(v-p)/p*100 for v, p in zip(va_vals, producer_vals)]
    
    ax2.plot(years, cons_diff, 's-', linewidth=2, markersize=5, label='Consumer vs Producer', color='#3498db')
    ax2.plot(years, va_diff, '^-', linewidth=2, markersize=5, label='Value-Added vs Producer', color='#2ecc71')
    ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    
    ax2.set_xlabel('Year', fontsize=11)
    ax2.set_ylabel('Difference from Producer (%)', fontsize=11)
    ax2.set_title('Methodological Differences Over Time\n(Relative to Producer Approach)', fontsize=12, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    
    # === PLOT 3: TOP COUNTRIES LATEST YEAR ===
    ax3 = plt.subplot(2, 3, 3)
    
    latest_year_idx = len(years) - 1
    latest_year = years[latest_year_idx]
    
    # Get top 10 countries by value-added approach
    latest_va_data = {country: values[latest_year_idx] 
                      for country, values in time_series_results['value_added_country'].items()
                      if len(values) > latest_year_idx}
    
    top_countries = sorted(latest_va_data.items(), key=lambda x: x[1], reverse=True)[:10]
    countries = [x[0] for x in top_countries]
    
    # Get data for all three approaches for these countries
    producer_vals_country = [time_series_results['producer_country'][c][latest_year_idx]/1e9 for c in countries]
    consumer_vals_country = [time_series_results['consumer_country'][c][latest_year_idx]/1e9 for c in countries]
    va_vals_country = [time_series_results['value_added_country'][c][latest_year_idx]/1e9 for c in countries]
    
    x = np.arange(len(countries))
    width = 0.25
    
    ax3.bar(x - width, producer_vals_country, width, label='Producer', alpha=0.8, color='#e74c3c')
    ax3.bar(x, consumer_vals_country, width, label='Consumer', alpha=0.8, color='#3498db')
    ax3.bar(x + width, va_vals_country, width, label='Value-Added', alpha=0.8, color='#2ecc71')
    
    ax3.set_xlabel('Country/Region', fontsize=11)
    ax3.set_ylabel('GHG Emissions (Gt CO₂-eq)', fontsize=11)
    ax3.set_title(f'Top 10 Emitters by Approach\n({latest_year})', fontsize=12, fontweight='bold')
    ax3.set_xticks(x)
    ax3.set_xticklabels(countries, rotation=45, fontsize=9)
    ax3.legend(fontsize=9)
    ax3.grid(True, alpha=0.3, axis='y')
    
    # === PLOT 4: CHINA, US, EU TRENDS ===
    ax4 = plt.subplot(2, 3, 4)
    
    key_regions = ['CN', 'US', 'DE']  # China, US, Germany as EU representative
    colors = ['#e74c3c', '#3498db', '#f39c12']
    
    for i, region in enumerate(key_regions):
        if region in time_series_results['value_added_country']:
            region_data = [val/1e9 for val in time_series_results['value_added_country'][region]]
            if len(region_data) == len(years):
                ax4.plot(years, region_data, 'o-', linewidth=2, markersize=4, 
                        label=f'{region}', color=colors[i])
    
    ax4.set_xlabel('Year', fontsize=11)
    ax4.set_ylabel('GHG Emissions (Gt CO₂-eq)', fontsize=11)
    ax4.set_title('Major Economies Trends\n(Value-Added Approach)', fontsize=12, fontweight='bold')
    ax4.legend(fontsize=10)
    ax4.grid(True, alpha=0.3)
    
    # === PLOT 5: GROWTH RATES ===
    ax5 = plt.subplot(2, 3, 5)
    
    if len(years) > 1:
        # Calculate year-over-year growth rates
        producer_growth = [((producer_vals[i]/producer_vals[i-1])-1)*100 
                          for i in range(1, len(producer_vals))]
        consumer_growth = [((consumer_vals[i]/consumer_vals[i-1])-1)*100 
                          for i in range(1, len(consumer_vals))]
        va_growth = [((va_vals[i]/va_vals[i-1])-1)*100 
                    for i in range(1, len(va_vals))]
        
        growth_years = years[1:]
        
        ax5.plot(growth_years, producer_growth, 'o-', linewidth=2, markersize=4, 
                label='Producer', color='#e74c3c', alpha=0.7)
        ax5.plot(growth_years, consumer_growth, 's-', linewidth=2, markersize=4, 
                label='Consumer', color='#3498db', alpha=0.7)
        ax5.plot(growth_years, va_growth, '^-', linewidth=2, markersize=4, 
                label='Value-Added', color='#2ecc71', alpha=0.7)
        
        ax5.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        ax5.set_xlabel('Year', fontsize=11)
        ax5.set_ylabel('Annual Growth Rate (%)', fontsize=11)
        ax5.set_title('Annual Growth Rates\n(Year-over-Year)', fontsize=12, fontweight='bold')
        ax5.legend(fontsize=9)
        ax5.grid(True, alpha=0.3)
    
    # === PLOT 6: EMISSIONS INTENSITY TRENDS ===
    ax6 = plt.subplot(2, 3, 6)
    
    # This would require GDP data - for now show total emissions per year
    total_emissions_by_year = producer_vals
    
    ax6.bar(years, total_emissions_by_year, alpha=0.7, color='#34495e', width=0.8)
    ax6.set_xlabel('Year', fontsize=11)
    ax6.set_ylabel('Total Emissions (Gt CO₂-eq)', fontsize=11)
    ax6.set_title('Total Global Emissions\n(Producer Approach)', fontsize=12, fontweight='bold')
    ax6.grid(True, alpha=0.3, axis='y')
    
    # Add trend line
    if len(years) > 2:
        z = np.polyfit(years, total_emissions_by_year, 1)
        p = np.poly1d(z)
        ax6.plot(years, p(years), "r--", alpha=0.8, linewidth=2)
        
        # Calculate and display trend
        annual_trend = z[0]
        ax6.text(0.05, 0.95, f'Trend: {annual_trend:+.2f} Gt/year', 
                transform=ax6.transAxes, fontsize=9, 
                bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # === SUMMARY STATISTICS ===
    print(f"\n=== SUMMARY STATISTICS ===")
    print(f"Time period: {years[0]}-{years[-1]} ({len(years)} years)")
    print(f"Latest year global emissions ({years[-1]}):")
    print(f"  Producer approach:    {producer_vals[-1]:.1f} Gt CO₂-eq")
    print(f"  Consumer approach:    {consumer_vals[-1]:.1f} Gt CO₂-eq")
    print(f"  Value-added approach: {va_vals[-1]:.1f} Gt CO₂-eq")
    
    if len(years) > 1:
        total_change = ((producer_vals[-1]/producer_vals[0])**(1/(len(years)-1)) - 1) * 100
        print(f"\nAverage annual growth rate: {total_change:+.1f}%")
        print(f"Total change {years[0]}-{years[-1]}: {((producer_vals[-1]/producer_vals[0])-1)*100:+.1f}%")
    
    # Data quality check
    mass_conservation_error = abs(consumer_vals[-1] - producer_vals[-1]) / producer_vals[-1] * 100
    print(f"\nMass conservation error (latest year): {mass_conservation_error:.3f}%")
    
    return fig

# Create the comprehensive visualization
fig = create_comprehensive_time_series_plots(time_series_results)
```

```python
def diagnose_emissions(year=2020):
    ixi_data = pymrio.parse_exiobase3(str(ixi_folder / f"IOT_{year}_ixi.zip"))
    air_emissions = ixi_data.air_emissions.F
    
    print(f"\n=== EMISSION DIAGNOSTICS FOR {year} ===")
    
    # Test different unit assumptions
    test_factors = {
        'Assuming kg': 1/1000,      # kg to tons
        'Assuming tonnes': 1,        # tonnes to tons
        'Assuming kt': 1000,         # kt to tons
        'Assuming Mt': 1e6           # Mt to tons
    }
    
    for assumption, factor in test_factors.items():
        total_co2_only = 0
        if 'CO2 - combustion - air' in air_emissions.index:
            co2_data = air_emissions.loc['CO2 - combustion - air'].values
            total_co2_only = co2_data.sum() * factor / 1e9  # Convert to Gt
        
        print(f"\n{assumption}:")
        print(f"  CO2 combustion only: {total_co2_only:.1f} Gt")
        print(f"  Expected (~33-36 Gt) ✓" if 30 < total_co2_only < 40 else "  Unlikely ✗")

diagnose_emissions(2020)
```
