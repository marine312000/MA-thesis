PRODUCER APPROACH

```python
def calculate_producer_responsibility(ixi_data, emission_categories):
    """
    Calculate producer-based responsibility.
    Units: tonnes CO2-eq per sector-region
    """
    # Get emissions data (EXIOBASE is in kilograms!)
    air_emissions_kg = ixi_data.air_emissions.F
    
    # Convert kg to tonnes (divide by 1000)
    air_emissions_tonnes = air_emissions_kg / 1000  # kg → tonnes
    
    
    # Check for missing categories
    missing_emissions = [em for em in emission_categories.keys() 
                        if em not in air_emissions_tonnes.index.tolist()]
    
    # Initialize results array
    n_sectors_regions = len(air_emissions_tonnes.columns)
    producer_emissions = np.zeros(n_sectors_regions)  # Will contain tonnes
    
    # Calculate total GHG emissions
    for emission_type, gwp_factor in emission_categories.items():
        if emission_type in air_emissions_tonnes.index:
            # Get emissions in tonnes
            emission_data_tonnes = air_emissions_tonnes.loc[emission_type].values
            
            # Convert to tonnes CO2-eq
            co2_equiv_tonnes = emission_data_tonnes * gwp_factor
            
            # Add to total
            producer_emissions += co2_equiv_tonnes
    
    return producer_emissions, missing_emissions  # Returns tonnes
```

CONSUMER APPRAOCH

```python
def calculate_consumer_responsibility(ixi_data, producer_emissions):
    """
    Calculate consumer-based responsibility following Piñero et al. (2019).
    
    Key insight: We must track which country consumes, not where production occurs.
    Formula: t' = f' × B × ŷ where ŷ is country-specific final demand
    Producer emissions in Gt
    """
    # Get dimensions
    regions = ixi_data.get_regions()
    sectors = ixi_data.get_sectors()
    
    # Calculate emission intensities (f' in Piñero)
    total_output = ixi_data.x.values.flatten()
   
    emission_intensity_t = np.divide(
    producer_emissions,   # Convert kg → tonnes
    total_output,
    out=np.zeros_like(producer_emissions),
    where=(total_output != 0)
)
# This gives you: tonnes CO2-eq per million EUR
    
    # Get Leontief inverse (B in Piñero)
    B = ixi_data.L.values
    
    # Get final demand matrix
    Y_full = ixi_data.Y
    
    # Extract final demand categories for detailed analysis
    fd_categories = Y_full.columns.get_level_values(1).unique()
    
    # Initialize results storage
    consumer_by_country = {}
    consumer_by_country_category = {}  # NEW: Keep category detail
    consumer_by_sector_region = {}
    
    # Calculate for each consuming country
    for region in regions:
        # Get ALL columns where this region is the consumer
        region_mask = Y_full.columns.get_level_values(0) == region
        Y_region_all = Y_full.loc[:, region_mask]
        
        # Store by category for detailed analysis
        consumer_by_country_category[region] = {}
        
        # Calculate for each final demand category
        for category in fd_categories:
            # Get specific category for this region
            category_mask = Y_region_all.columns.get_level_values(1) == category
            Y_region_category = Y_region_all.loc[:, category_mask].sum(axis=1).values
            
            # Apply Piñero formula: t' = f' × B × ŷ
            # Step 1: B × ŷ (total requirements)
            total_requirements = B @ Y_region_category
            
            # Step 2: f' × (B × ŷ) (emissions)
            emissions_category = emission_intensity_t * total_requirements
            
            # Store category-specific results
            consumer_by_country_category[region][category] = emissions_category.sum()
        
        # Also calculate total for the country (all categories)
        Y_region_total = Y_region_all.sum(axis=1).values
        total_requirements_all = B @ Y_region_total
        emissions_total = emission_intensity_t * total_requirements_all
        
        # Store results
        consumer_by_country[region] = emissions_total.sum()
        consumer_by_sector_region[region] = pd.Series(
            emissions_total,
            index=ixi_data.x.index
        )
    
    return consumer_by_country, consumer_by_sector_region, consumer_by_country_category
```

VALUE ADDED APPRAOCH

```python
def calculate_value_added_responsibility(ixi_data, producer_emissions, 
                                       value_added_components):
    """
    Calculate value-added-based responsibility following Piñero et al. (2019).
    
    Formula: h = v̂ × B × t where t is consumer-based emissions
    But we implement directly as: h = f × diag(v) × B × ŷ for efficiency
    """
    # Get dimensions
    regions = ixi_data.get_regions()
    
    # Get value-added data
    factor_inputs = ixi_data.factor_inputs.F
    available_va_components = [comp for comp in value_added_components 
                              if comp in factor_inputs.index]
    missing_va_components = list(set(value_added_components) - set(available_va_components))
    
    # Calculate total value-added
    total_value_added = factor_inputs.loc[available_va_components].sum(axis=0).values
    
    # Calculate value-added coefficients (v in Piñero)
    total_output = ixi_data.x.values.flatten()
    va_coefficients = np.divide(
        total_value_added,
        total_output,
        out=np.zeros_like(total_value_added),
        where=(total_output != 0)
    )
    
    # Clean coefficients
    va_coefficients = np.clip(va_coefficients, 0, 1)
    
    # Calculate emission intensities (f in Piñero)
    emission_intensity_t = np.divide(
        producer_emissions,
        total_output,
        out=np.zeros_like(producer_emissions),
        where=(total_output != 0)
    )
    
    # Get system matrices
    B = ixi_data.L.values
    Y_full = ixi_data.Y
    
    # Create diagonal value-added matrix
    V_diag = np.diag(va_coefficients)
    
    # Pre-calculate V × B for efficiency
    VB = V_diag @ B
    
    # Initialize results
    va_by_country_raw = {}
    va_by_sector_region = {}
    va_by_country_category = {}  # NEW: Keep category detail
    
    # Get final demand categories
    fd_categories = Y_full.columns.get_level_values(1).unique()
    
    for region in regions:
        # Get final demand for this consuming region
        region_mask = Y_full.columns.get_level_values(0) == region
        Y_region_all = Y_full.loc[:, region_mask]
        
        # By category
        va_by_country_category[region] = {}
        for category in fd_categories:
            category_mask = Y_region_all.columns.get_level_values(1) == category
            Y_region_category = Y_region_all.loc[:, category_mask].sum(axis=1).values
            
            # Apply formula: h = f × V × B × ŷ
            va_weighted_requirements = VB @ Y_region_category
            emissions_va = emission_intensity_t * va_weighted_requirements
            
            va_by_country_category[region][category] = emissions_va.sum()
        
        # Total for country
        Y_region_total = Y_region_all.sum(axis=1).values
        va_weighted_requirements_total = VB @ Y_region_total
        emissions_va_total = emission_intensity_t * va_weighted_requirements_total
        
        va_by_country_raw[region] = emissions_va_total.sum()
        va_by_sector_region[region] = pd.Series(
            emissions_va_total,
            index=ixi_data.x.index
        )
    
    # Normalize for mass conservation (as in Piñero)
    # First get consumer total (need to recalculate if not passed)
    consumer_total = sum(calculate_consumer_responsibility(ixi_data, producer_emissions)[0].values())
    va_total_raw = sum(va_by_country_raw.values())
    normalization_factor = consumer_total / va_total_raw if va_total_raw > 0 else 1.0
    
    # Apply normalization to all results
    va_by_country = {}
    for region in regions:
        va_by_country[region] = va_by_country_raw[region] * normalization_factor
        va_by_sector_region[region] = va_by_sector_region[region] * normalization_factor
        for category in va_by_country_category[region]:
            va_by_country_category[region][category] *= normalization_factor
    
    return va_by_country, va_by_sector_region, va_by_country_category, missing_va_components
```

MAIN ANALYSIS

```python
def analyze_single_year(year, exiobase_path, emission_categories, 
                       value_added_components, verbose=True):
    """
    Complete three-approach analysis for a single year.
    Now includes final demand category breakdown.
    """
    start_time = time.time()
    
    if verbose:
        print(f"\nAnalyzing year {year}")
        print("-" * 40)
    
    # Load data
    ixi_data = pymrio.parse_exiobase3(
        str(exiobase_path / f"IOT_{year}_ixi.zip")
    )
    
    # Calculate system matrices ONCE (remove redundancy)
    ixi_data.calc_system()
    
    # Get dimensions ONCE
    regions = ixi_data.get_regions()
    sectors = ixi_data.get_sectors()
    n_sectors = len(sectors)
    
    if verbose:
        print(f"Data loaded: {len(regions)} regions, {len(sectors)} sectors")
    
    # 1. Producer-based responsibility
    producer_emissions, missing_emissions = calculate_producer_responsibility(
        ixi_data, emission_categories
    )
    
    # Aggregate by country (more efficient method)
    producer_by_country = {}
    producer_by_sector = pd.Series(producer_emissions, index=ixi_data.x.index)
    
    # Group by first level of index (country)
    for region in regions:
        region_mask = producer_by_sector.index.get_level_values(0) == region
        producer_by_country[region] = producer_by_sector[region_mask].sum()
    
    if verbose:
        total_producer = sum(producer_by_country.values())
        print(f"\nTotal emissions (million tonnes CO2-eq):")
        print(f"Producer-based: {total_producer/1e6:.1f}")

        print("DEBUG - Producer emissions check:")
        print(f"  Type: {type(producer_emissions)}")
        print(f"  Shape: {producer_emissions.shape}")
        print(f"  Sum: {producer_emissions.sum()/1e9:.1f} Gt")
        print(f"  Max: {producer_emissions.max()/1e9:.1f} Gt")
        print(f"  Expected magnitude: ~43 Gt")
    
    # 2. Consumer-based responsibility (with category detail)
    consumer_results = calculate_consumer_responsibility(ixi_data, producer_emissions)
    consumer_by_country = consumer_results[0]
    consumer_by_sector = consumer_results[1]
    consumer_by_category = consumer_results[2]
    
    if verbose:
        total_consumer = sum(consumer_by_country.values())
        print(f"Consumer-based: {total_consumer/1e6:.1f}")
    
    # 3. Value-added-based responsibility (with category detail)
    va_results = calculate_value_added_responsibility(
        ixi_data, producer_emissions, value_added_components
    )
    va_by_country = va_results[0]
    va_by_sector = va_results[1]
    va_by_category = va_results[2]
    missing_va = va_results[3]
    
    if verbose:
        total_va = sum(va_by_country.values())
        print(f"Value-added-based: {total_va/1e6:.1f}")
    
    # Processing time
    processing_time = time.time() - start_time
    if verbose:
        print(f"Processing time: {processing_time:.1f} seconds")
    
    # Compile results (now includes category breakdowns)
    results = {
        'year': year,
        # Country totals
        'producer_by_country': pd.Series(producer_by_country),
        'consumer_by_country': pd.Series(consumer_by_country),
        'value_added_by_country': pd.Series(va_by_country),
        # Sector details
        'producer_by_sector': producer_by_sector,
        'consumer_by_sector': consumer_by_sector,
        'value_added_by_sector': va_by_sector,
        # NEW: Category breakdowns
        'consumer_by_category': consumer_by_category,
        'value_added_by_category': va_by_category,
        # Metadata
        'missing_emissions': missing_emissions,
        'missing_va_components': missing_va,
        'processing_time': processing_time,
        'regions': regions,
        'sectors': sectors
    }
    
    return results
```

COMPARISION ANALYSIS

```python
def create_comparison_dataframe(results):
    """Create a dataframe comparing all three approaches."""
    
    comparison = pd.DataFrame({
        'Producer': results['producer_by_country'],
        'Consumer': results['consumer_by_country'],
        'Value_Added': results['value_added_by_country']
    })
    
    # Add derived metrics
    comparison['Prod_to_Cons'] = comparison['Consumer'] - comparison['Producer']
    comparison['Cons_to_VA'] = comparison['Value_Added'] - comparison['Consumer']
    comparison['Net_Transfer'] = comparison['Consumer'] - comparison['Producer']
    
    # Sort by total emissions
    comparison['Total'] = comparison[['Producer', 'Consumer', 'Value_Added']].sum(axis=1)
    comparison = comparison.sort_values('Total', ascending=False)
    
    return comparison

# Create comparison
comparison_2020 = create_comparison_dataframe(results_2020)

# Display key insights
print("\nKEY INSIGHTS:")
print("-" * 40)

# Net exporters (Producer > Consumer)
net_exporters = comparison_2020[comparison_2020['Net_Transfer'] < 0].head(5)
print("\nTop 5 net exporters of emissions:")
for country in net_exporters.index:
    transfer = comparison_2020.loc[country, 'Net_Transfer'] / 1e9 # tonnes → Gt
    print(f"  {country}: {transfer:.2f} Gt")

# Net importers (Consumer > Producer)
net_importers = comparison_2020[comparison_2020['Net_Transfer'] > 0].head(5)
print("\nTop 5 net importers of emissions:")
for country in net_importers.index:
    transfer = comparison_2020.loc[country, 'Net_Transfer'] / 1e9
    print(f"  {country}: +{transfer:.2f} Gt")
```

TIME SERIES ANALYSIS

```python
def run_time_series_analysis(years, exiobase_path, emission_categories,
                           value_added_components, cache_file=None):
    """
    Run analysis for multiple years with caching support.
    
    Parameters:
    -----------
    years : list
        List of years to analyze
    cache_file : Path, optional
        Path to cache file for saving/loading results
    
    Returns:
    --------
    time_series_results : dict
        Results for all years
    """
    # Check for cached results
    if cache_file and cache_file.exists():
        print(f"Loading cached results from {cache_file}")
        with open(cache_file, 'rb') as f:
            return pickle.load(f)
    
    # Initialize storage
    time_series_results = {
        'years': [],
        'producer_by_country': {},
        'consumer_by_country': {},
        'value_added_by_country': {},
        'global_totals': {'producer': [], 'consumer': [], 'value_added': []},
        'processing_times': []
    }
    
    # Get list of all countries (from first year)
    all_countries = set()
    
    # Run analysis for each year
    for i, year in enumerate(years):
        print(f"\nProcessing year {i+1}/{len(years)}: {year}")
        
        try:
            results = analyze_single_year(
                year=year,
                exiobase_path=exiobase_path,
                emission_categories=emission_categories,
                value_added_components=value_added_components,
                verbose=False
            )
            
            # Store results
            time_series_results['years'].append(year)
            
            # Store country-level data
            for country in results['producer_by_country'].index:
                all_countries.add(country)
                
                for approach in ['producer', 'consumer', 'value_added']:
                    key = f'{approach}_by_country'
                    if country not in time_series_results[key]:
                        time_series_results[key][country] = []
                    
                    time_series_results[key][country].append(
                        results[key].get(country, 0)
                    )
            
            # Store global totals
            for approach in ['producer', 'consumer', 'value_added']:
                total = results[f'{approach}_by_country'].sum()
                time_series_results['global_totals'][approach].append(total)
            
            time_series_results['processing_times'].append(results['processing_time'])
            
            print(f"  Completed in {results['processing_time']:.1f}s")
            
        except Exception as e:
            print(f"  Error processing {year}: {e}")
    
    # Convert to arrays for easier plotting
    for approach in ['producer', 'consumer', 'value_added']:
        key = f'{approach}_by_country'
        for country in time_series_results[key]:
            time_series_results[key][country] = np.array(
                time_series_results[key][country]
            )
    
    # Save cache if requested
    if cache_file:
        print(f"\nSaving results to {cache_file}")
        cache_file.parent.mkdir(exist_ok=True)
        with open(cache_file, 'wb') as f:
            pickle.dump(time_series_results, f)
    
    return time_series_results
```

VISUALIZATION FUNCTION

```python
def plot_global_trends(time_series_results):
    """Plot global emission trends for all three approaches."""
    
    years = time_series_results['years']

    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Left panel: Absolute values
    for approach, label in [('producer', 'Producer'), 
                           ('consumer', 'Consumer'), 
                           ('value_added', 'Value-Added')]:
        values = np.array(time_series_results['global_totals'][approach]) / 1e9
        ax1.plot(years, values, 'o-', linewidth=2, markersize=5, label=label)
    
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Global GHG Emissions (Gt CO2-eq)')
    ax1.set_title('Global Emissions by Allocation Method')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Right panel: Growth rates
    for approach, label in [('producer', 'Producer'), 
                           ('consumer', 'Consumer'), 
                           ('value_added', 'Value-Added')]:
        values = np.array(time_series_results['global_totals'][approach]) / 1e9
        growth = (values[1:] - values[:-1]) / values[:-1] * 100
        ax2.plot(years[1:], growth, 'o-', linewidth=2, markersize=5, label=label)
    
    ax2.axhline(0, color='black', linestyle='--', alpha=0.5)
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Annual Growth Rate (%)')
    ax2.set_title('Year-over-Year Growth')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    return fig

# Plot global trends
fig_trends = plot_global_trends(time_series_results)
plt.show()
```

PER CAPITA ANALYSIS

```python

def calculate_per_capita_emissions(results, population):
    """Calculate per capita emissions for all approaches."""
    
    per_capita = pd.DataFrame()
    
    for approach in ['producer', 'consumer', 'value_added']:
        country_data = results[f'{approach}_by_country']
        
        # Calculate per capita (tons CO2-eq per person) - data is in kg, convert to tonnes
        per_capita[approach] = (country_data) / population  
        
    # Remove countries with missing population data
    per_capita = per_capita.dropna()
    
    return per_capita

# Calculate per capita for 2020 (using 2022 population as proxy)
if len(population_2022) > 0:
    per_capita_2020 = calculate_per_capita_emissions(results_2020, population_2022)
    
    # Display top per capita emitters
    print("\nTOP 10 PER CAPITA EMITTERS (2020):")
    print("-" * 60)
    
    for approach in ['producer', 'consumer', 'value_added']:
        print(f"\n{approach.upper()} APPROACH (tons CO2-eq/person):")
        top10 = per_capita_2020[approach].nlargest(10)
        for i, (country, emissions) in enumerate(top10.items(), 1):
            print(f"{i:2d}. {country}: {emissions:6.1f} tonnes CO2-eq/person")
```

SECTORAL ANALYSIS

```python
def analyze_sectors_global(results):
    """Analyze emissions by sector globally."""
    
    # Get sector names from first country's data
    first_country = list(results['consumer_by_sector'].keys())[0]
    sectors_index = results['consumer_by_sector'][first_country].index
    
    # Aggregate across all countries
    producer_by_sector = results['producer_by_sector'].copy()
    
    # For consumer and value-added, sum across countries
    consumer_by_sector_global = pd.Series(0, index=sectors_index)
    va_by_sector_global = pd.Series(0, index=sectors_index)
    
    for country in results['consumer_by_sector']:
        consumer_by_sector_global += results['consumer_by_sector'][country]
        va_by_sector_global += results['value_added_by_sector'][country]
    
    # Create sector comparison dataframe
    sector_comparison = pd.DataFrame({
        'Producer': producer_by_sector,
        'Consumer': consumer_by_sector_global,
        'Value_Added': va_by_sector_global
    })
    
    # Add sector names for clarity
    sector_comparison['Country'] = [idx[0] for idx in sector_comparison.index]
    sector_comparison['Sector'] = [idx[1] for idx in sector_comparison.index]
    
    return sector_comparison

# Analyze sectors
sector_analysis = analyze_sectors_global(results_2020)

# Find top emitting sectors
print("\nTOP EMITTING SECTORS GLOBALLY (2020):")
print("-" * 80)

# Group by sector name only (aggregate across countries)
sector_totals = sector_analysis.groupby('Sector')[['Producer', 'Consumer', 'Value_Added']].sum()
sector_totals['Total'] = sector_totals.sum(axis=1)
sector_totals = sector_totals.sort_values('Total', ascending=False)

print("\nTop 15 sectors by total emissions:")
for i, (sector, row) in enumerate(sector_totals.head(15).iterrows(), 1):
    print(f"\n{i:2d}. {sector[:60]}")
    print(f"    Producer: {row['Producer']/1e9:6.2f} Gt")
    print(f"    Consumer: {row['Consumer']/1e9:6.2f} Gt")
    print(f"    Value-Added: {row['Value_Added']/1e9:6.2f} Gt")
```

```python
def analyze_country_sectors(results, country_code):
    """Detailed analysis for a specific country."""
    
    print(f"\nDETAILED ANALYSIS FOR {country_code}")
    print("=" * 60)
    
    # Country totals
    print(f"\nTotal emissions:")
    print(f"  Producer: {results['producer_by_country'][country_code]/1e9:.2f} Gt")
    print(f"  Consumer: {results['consumer_by_country'][country_code]/1e9:.2f} Gt")
    print(f"  Value-Added: {results['value_added_by_country'][country_code]/1e9:.2f} Gt")
    
    # Get sectors for this country
    country_sectors = [(idx[0], idx[1]) for idx in results['producer_by_sector'].index 
                       if idx[0] == country_code]
    
    # Producer emissions by sector in this country
    print(f"\nTop 10 producing sectors in {country_code}:")
    producer_sectors = results['producer_by_sector'][
        results['producer_by_sector'].index.get_level_values(0) == country_code
    ].nlargest(10)
    
    for i, (idx, value) in enumerate(producer_sectors.items(), 1):
        sector = idx[1]
        print(f"{i:2d}. {sector[:50]}: {value/1e9:.1f} Mt CO2-eq")
    
    # Consumer emissions by sector
    print(f"\nTop 10 consumed sectors by {country_code}:")
    consumer_sectors = results['consumer_by_sector'][country_code].nlargest(10)
    
    for i, (idx, value) in enumerate(consumer_sectors.items(), 1):
        sector = idx[1]
        origin = idx[0]
        print(f"{i:2d}. {sector[:40]} (from {origin}): {value/1e9:.1f} Mt CO2-eq")

# Analyze specific countries
for country in ['CN', 'US', 'DE', 'IN']:
    if country in results_2020['producer_by_country'].index:
        analyze_country_sectors(results_2020, country)
```

```python
# EXIOBASE Factor Inputs Investigation
import pymrio
import pandas as pd

def investigate_factor_inputs(exiobase_path, year=2020):
    """
    Investigate all available factor inputs in EXIOBASE 3
    """
    print("=== EXIOBASE FACTOR INPUTS INVESTIGATION ===")
    print(f"Loading EXIOBASE {year}...")
    
    # Load EXIOBASE data
    ixi_data = pymrio.parse_exiobase3(
        str(exiobase_path / f"IOT_{year}_ixi.zip")
    )
    
    print(f"Data loaded successfully for {year}")
    print("-" * 60)
    
    # Get factor inputs
    factor_inputs = ixi_data.factor_inputs.F
    
    print("1. ALL FACTOR INPUTS AVAILABLE:")
    print(f"   Total number: {len(factor_inputs.index)}")
    print()
    
    for i, factor in enumerate(factor_inputs.index, 1):
        # Calculate some basic stats
        values = factor_inputs.loc[factor]
        total = values.sum()
        max_val = values.max()
        non_zero = (values > 0).sum()
        
        print(f"{i:2d}. {factor}")
        print(f"    Total: {total:>15,.0f}")
        print(f"    Max value: {max_val:>12,.0f}")
        print(f"    Non-zero entries: {non_zero:>6d} of {len(values)}")
        print()
    
    print("-" * 60)
    
    # 2. Categorize them
    print("2. FACTOR INPUTS BY CATEGORY:")
    
    # Compensation (labor)
    compensation = [f for f in factor_inputs.index if 'Compensation' in f]
    print(f"\n📊 COMPENSATION/LABOR ({len(compensation)} items):")
    for comp in compensation:
        print(f"   • {comp}")
    
    # Operating surplus (capital returns)
    operating = [f for f in factor_inputs.index if 'Operating surplus' in f]
    print(f"\n💰 OPERATING SURPLUS/CAPITAL ({len(operating)} items):")
    for op in operating:
        print(f"   • {op}")
    
    # Taxes
    taxes = [f for f in factor_inputs.index if 'taxes' in f.lower() or 'tax' in f.lower()]
    print(f"\n🏛️  TAXES ({len(taxes)} items):")
    for tax in taxes:
        print(f"   • {tax}")
    
    # Other
    all_categorized = compensation + operating + taxes
    other = [f for f in factor_inputs.index if f not in all_categorized]
    if other:
        print(f"\n❓ OTHER ({len(other)} items):")
        for oth in other:
            print(f"   • {oth}")
    
    print("-" * 60)
    
    # 3. Suggest value-added components
    print("3. SUGGESTED VALUE-ADDED COMPONENTS FOR YOUR ANALYSIS:")
    
    # Standard value-added = compensation + operating surplus + taxes on production
    suggested = []
    
    # Add all compensation
    suggested.extend(compensation)
    
    # Add operating surplus (these are profits/capital returns)
    suggested.extend(operating)
    
    # Add taxes on production (not on products)
    production_taxes = [f for f in taxes if 'production' in f.lower()]
    suggested.extend(production_taxes)
    
    print(f"\n✅ RECOMMENDED SET ({len(suggested)} components):")
    for i, comp in enumerate(suggested, 1):
        total_val = factor_inputs.loc[comp].sum()
        print(f"{i:2d}. {comp}")
        print(f"    Total value: {total_val:>15,.0f} million EUR")
    
    # Calculate totals
    total_va = factor_inputs.loc[suggested].sum().sum()
    total_all = factor_inputs.sum().sum()
    coverage = (total_va / total_all) * 100
    
    print(f"\n📈 COVERAGE ANALYSIS:")
    print(f"   Recommended VA total: {total_va:>15,.0f} million EUR")
    print(f"   All factor inputs total: {total_all:>15,.0f} million EUR")
    print(f"   Coverage: {coverage:.1f}% of total factor inputs")
    
    print("-" * 60)
    
    # 4. Return the list for easy copying
    print("4. PYTHON CODE FOR YOUR ANALYSIS:")
    print("\n# Copy this into your code:")
    print("VALUE_ADDED_COMPONENTS = [")
    for comp in suggested:
        print(f'    "{comp}",')
    print("]")
    
    return {
        'all_factors': list(factor_inputs.index),
        'suggested_va': suggested,
        'compensation': compensation,
        'operating_surplus': operating,
        'taxes': taxes,
        'other': other,
        'coverage_percent': coverage
    }

# Run the investigation
if __name__ == "__main__":
    # Your EXIOBASE path
    from pathlib import Path
    EXIOBASE_PATH = Path(r"C:\Users\Marine Riesterer\OneDrive\Desktop\MA Arbeit\Exiobase3_ixi_latest")
    
    results = investigate_factor_inputs(EXIOBASE_PATH, year=2020)
    
    print("\n" + "="*60)
    print("INVESTIGATION COMPLETE!")
    print(f"Found {len(results['all_factors'])} total factor inputs")
    print(f"Recommended {len(results['suggested_va'])} for value-added analysis")
    print(f"Coverage: {results['coverage_percent']:.1f}%")
    print("="*60)
```
